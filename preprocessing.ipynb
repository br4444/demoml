# Str -> int
- LabelEncoder
- get_dummies
- OneHotEncoder
- ...

```python
# exemple de dataset
list_test_encoder = ['croissant',
                     'pain au chocolat',
                     'Pain aux raisins',
                     'pain au chocolat',
                     'croissant',
                     'Pain aux raisins']
list_jout = ['mercredi', 'lundi', 'mercredi', 'lundi', 'mercredi', 'lundi']
df_test_encoder = pd.DataFrame({'viennoiserie': list_test_encoder, 'jour': list_jout})
df_test_encoder
```

LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
encoder1 = LabelEncoder()

encoder1.fit_transform(df_test_encoder['viennoiserie'])
```

get_dummies

```python
#pd.get_dummies(df_test_encoder, columns=['viennoiserie'])
pd.get_dummies(df_test_encoder, columns=['viennoiserie', 'jour'])
```

OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
encoder2 = OneHotEncoder()

encoded = encoder2.fit_transform(df_test_encoder[['viennoiserie']]).toarray()
pd.DataFrame(encoded, columns=encoder2.get_feature_names_out(['viennoiserie']))
```
